```python
import pandas as pd
```

```python
omni_channel = pd.read_csv("HS_MODEL_V2.csv")
```

```python
masked_columns = omni_channel.columns.to_list()
```

```python
masked_columns[0:3] = ['CUSTOMER_ID','STORE_NO','STORE_TYPE']
```

```python
masked_columns_replaced = []
for x in masked_columns:
    x_rep = x.replace("HSONU","WEEKEND")
    masked_columns_replaced.append(x_rep)
```

```python
masked_columns_replaced_1 = []
for i,x in enumerate(masked_columns_replaced[48:67]):
    x_rep_1 = 'CAT_' + str(i+1) + '_VISIT_PEN'
    masked_columns_replaced_1.append(x_rep_1)
```

```python
masked_columns_replaced_2 = []
for i,x in enumerate(masked_columns_replaced[67:86]):
    x_rep_2 = 'CAT_' + str(i+1) + '_REVENUE_PEN'
    masked_columns_replaced_2.append(x_rep_2)
```

```python
masked_columns_replaced[48:67] = masked_columns_replaced_1
masked_columns_replaced[67:86] = masked_columns_replaced_2
```

```python
omni_channel.columns = masked_columns_replaced
```

```python
omni_channel.head()
```

```python
omni_channel['GENDER'] = omni_channel['GENDER'].map({'ERKEK':'Male','KADIN':'Female','YOK':None})
```

```python
omni_channel['GENDER'].value_counts()
```

```python
omni_channel['GENDER'].isna().sum()
```

```python
omni_channel.head()
```

```python
omni_channel['AGE_GRP'] = omni_channel['AGE_GRP'].apply(lambda x: None if x == '0) Unknown' else x.replace(x[0:3],''))
```

```python
omni_channel['AGE_GRP'].value_counts()
```

```python
omni_channel.iloc[:,26:40]
```

```python
status_cols = ['STATUS_L12M','STATUS_L9M','STATUS_L6M','STATUS_L3M']

for col in status_cols:
    omni_channel[col] = omni_channel[col].map({'MULTI':'Multi','SINGLE':'Single','YOK':None})
```

```python
rev_grp_cols = ['REV_GRP_L12M','REV_GRP_L9M','REV_GRP_L6M','REV_GRP_L3M']

for col in rev_grp_cols:
    omni_channel[col] = omni_channel[col].apply(lambda x: None if x == '0 ) YOK' else x.replace(x[0:4],''))
```

```python
visit_grp_cols = ['VISIT_GRP_L12M','VISIT_GRP_L9M','VISIT_GRP_L6M','VISIT_GRP_L3M','RECENCY_GRP','TENURE_GRP']

for col in visit_grp_cols:
    omni_channel[col] = omni_channel[col].apply(lambda x: None if x == '0) YOK' else x.replace(x[0:3],''))
```
